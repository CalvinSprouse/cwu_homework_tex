```python
# imports
import os

from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rc
from rich import print
```

```python
# matplotlib setup

# matplotlib inline
%matplotlib inline

# matplotlib rc config
# increase the font size
rc('font', **{'family': 'serif', 'serif': ['Times'], 'size': 12})
rc('text', usetex=True)

# figure save location
figure_dir = Path(".figures")
os.makedirs(figure_dir, exist_ok=True)
```

```python
# define the person / sibling dataset
sibling_count_list = [1, 2, 0, 3, 0, 1, 5, 1, 2, 4]

# create a histogram from the list of numbers in matplotlib
# create a figure
fig, ax = plt.subplots()

# plot the histogram
# make it a wire histogram
ax.hist(
    sibling_count_list,
    bins=range(7),
    align="left",
    color="lightblue",
    edgecolor="black",
    linewidth=1,
    density=False,
    )

# change the y-axis to display whole integers only
ax.set_yticks(range(0, 4))

# set the axis labels
ax.set_xlabel("Number of Siblings")
ax.set_ylabel("Frequency")
ax.set_title("A distribution of sibling count from 10 people")

# save the figure
fig.savefig(figure_dir / "sibling_count_histogram.pdf", bbox_inches="tight")
```

```python
# define the sodium level dataset
sodium_level_list = [136.1, 141.2, 140.3, 139.9, 142.1, 146.7, 134.8, 135.5, 144.4]

# the units on each entry of the sodium level list are mEq/L

# create a box plot from the sodium level list in matplotlib
# create a figure
fig, ax = plt.subplots(figsize=(5, 6))

# plot the box plot
ax.boxplot(
    sodium_level_list,
    patch_artist=True,
    boxprops=dict(facecolor="lightblue", color="black", linewidth=1.2),
    medianprops=dict(color="black", linewidth=1.2),
    whiskerprops=dict(color="black", linewidth=1.2),
    capprops=dict(color="black", linewidth=1.2),
    flierprops=dict(marker="o", markerfacecolor="black", markersize=5),
    widths=0.7
    )

# set the axis labels
ax.set_ylabel("Sodium Level [mEq/L]")
ax.set_title("The sodium level of 9 patients after drug administration")

# remove the horizontal ticks and label
ax.tick_params(axis="x", which="both", bottom=False, top=False)
ax.set_xticklabels([])

# save the figure
fig.savefig(figure_dir / "sodium_level_boxplot.pdf", bbox_inches="tight")
```

```python
# define the age dataset
age_list = [5, 2, 18, 15, 12, 20]

# create a box plot from the age_list
# create a figure
fig, ax = plt.subplots(figsize=(5, 6))

# plot the box plot
ax.boxplot(
    age_list,
    patch_artist=True,
    boxprops=dict(facecolor="lightblue", color="black", linewidth=1.2),
    medianprops=dict(color="black", linewidth=1.2),
    whiskerprops=dict(color="black", linewidth=1.2),
    capprops=dict(color="black", linewidth=1.2),
    flierprops=dict(marker="o", markerfacecolor="black", markersize=5),
    widths=0.7
    )

# set the axis labels
ax.set_ylabel("Age [years]")
ax.set_title("The age of 6 children in a family")

# change the y-axis to represent whole integers only in steps of 4
ax.set_yticks(range(0, 21, 2))

# remove the horizontal ticks and label
ax.tick_params(axis="x", which="both", bottom=False, top=False)
ax.set_xticklabels([])

# save the figure
fig.savefig(figure_dir / "age_boxplot.pdf", bbox_inches="tight")
```

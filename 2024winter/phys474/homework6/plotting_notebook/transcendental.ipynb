## Define and solve transcendental equations

```python
# define the lhs equation lhs = - cot(z)
def lhs(z):
    return -1 / np.tan(z)

# define the rhs equation sqrt(4/z^2 - 1)
def rhs(z):
    return np.sqrt((4 / z**2) - 1)
```

```python
# define a range of z values in multiples of pi
z = np.linspace(-2, 2, 1000) * np.pi/2

# define formatter to print the x axis in fractions of pi
def format_func(value, tick_number):
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\frac{\pi}{2}$"
    elif N == -1:
        return r"$-\frac{\pi}{2}$"
    elif N == 2:
        return r"$\pi$"
    elif N == -2:
        return r"$-\pi$"
    else:
        return r"$\frac{" + str(N) + r"\pi}{2}$"

# plot the lhs and rhs
fig, ax = plt.subplots()

# plot lhs and rhs
ax.plot(z, lhs(z), label=r"$-\cot(z)$", color="blue")
ax.plot(z, rhs(z), label=r"$\sqrt{\frac{4}{z^2} - 1}$", color="orange")

# set the x-axis ticks to be in multiples of pi
ax.set_xticks(np.arange(-2, 3) * np.pi / 2)
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))

# set y-axis limits
ax.set_ylim(-10, 10)

# set the x and y labels
ax.set_xlabel(r"$z$")
ax.set_ylabel(r"value of transcendental equation")
ax.set_title(r"Solutions of the transcendental equation $-cot(z) = \sqrt{\frac{4}{z^2} - 1}$")

# plot a vertical line at pi
ax.axvline(np.pi, color="black", linestyle="--")

# add a legend
ax.legend()

# save the figure
fig.savefig(figure_dir / "transcendental_equation.pdf")
```
